```python
#dropping columns that aren't helpful

cancerdata = cancerdata.drop(columns=['STDs: Time since last diagnosis','STDs: Time since first diagnosis'])
cancerdata
```

```python
#assigning columns to features and targets as needed to train our model accurately

features= cancerdata.drop(columns = ['Dx:Cancer'])

target = cancerdata['Dx:Cancer']
```

```python
#train test split

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```
